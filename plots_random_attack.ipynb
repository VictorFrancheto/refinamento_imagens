```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

def set_fallback_cm_fonts():
    plt.rcParams.update({
        "text.usetex": True,
        "font.family": "serif",
        "font.serif": ["DejaVu Serif"],
        "mathtext.fontset": "cm",
        "mathtext.rm": "serif",
        "mathtext.it": "serif:italic",
        "mathtext.bf": "serif:bold",
    })

def load_csv_attack(N, C, mode="random"):
    fname = f"./data_random_medias_random/{mode}_{N}_{C}.csv"
    if not os.path.exists(fname):
        print(f"[AVISO] Arquivo no encontrado: {fname}")
        return pd.DataFrame()
    df = pd.read_csv(fname, sep=',', header=0)
    return df.drop(columns=["node1", "node2"], errors="ignore")

def plot_m1_subplot(ax, N, show_legend=False):
    set_fallback_cm_fonts()

    # Paleta
    cmap = plt.cm.get_cmap('winter')
    styles = {
        'C=1': {'color': cmap(1.0), 'linestyle': '-'},    # slido
        'C=2': {'color': cmap(0.8), 'linestyle': '--'},   # tracejado
        'C=3': {'color': '#6BAED6', 'linestyle': ':'},    # pontilhado
        'C=N': {'color': '#2171B5', 'linestyle': '-.'},   # trao-ponto
    }

    # Carregamento dos dados
    C1_df = load_csv_attack(N, 1)
    C2_df = load_csv_attack(N, 2)
    C3_df = load_csv_attack(N, 3)
    CN_df = load_csv_attack(N, N)

    ax.set_axisbelow(True)
    ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

    # Plots
    ax.plot(C1_df["p"], C1_df["m1"], label=r"$C=1$", **styles['C=1'], linewidth=3.0)
    ax.plot(C2_df["p"], C2_df["m1"], label=r"$C=2$", **styles['C=2'], linewidth=3.0)
    ax.plot(C3_df["p"], C3_df["m1"], label=r"$C=3$", **styles['C=3'], linewidth=3.0)
    ax.plot(CN_df["p"], CN_df["m1"], label=r"$C=N$", **styles['C=N'], linewidth=3.0)

    # Eixos
    ax.set_xlim(left=0)
    ax.set_ylim(bottom=0)
    ax.set_xlabel(r"$p$", fontsize=18)
    ax.set_ylabel(r"$N_{G}/N$" if N in (300, 1000) else "", fontsize=18)
    ax.tick_params(labelsize=14, direction='in')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda val, _: '' if val == 0 else f'{val:g}'))
    ax.grid(False)

    # Legenda ou caixa de texto
    if show_legend:
        ax.legend(
            fontsize=14,
            title=rf"$N={N}$",
            title_fontsize=14,
            frameon=True,
            fancybox=True,
            edgecolor='black',
            loc='upper right'
        )
    else:
        ax.text(
            0.95, 0.95, rf"$N={N}$",
            transform=ax.transAxes,
            fontsize=14,
            ha='right', va='top',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black')
        )


def plot_m1_multi_layout():
    Ns = [300, 500, 600, 1000, 2000, 3000]
    fig, axs = plt.subplots(2, 3, figsize=(18, 9))

    for ax, N in zip(axs.flat, Ns):
        plot_m1_subplot(ax, N, show_legend=(N == 300))

    fig.tight_layout()
    #fig.savefig("attack_random_spp-quantum_.png", dpi=300, bbox_inches='tight', transparent=False)
    plt.show()


if __name__ == "__main__":
    plot_m1_multi_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assumindo que essas funes j existem no seu projeto:
# - set_fallback_cm_fonts()
# - load_csv_attack(N, C, mode)

# ---------------------------
# Limpeza
# ---------------------------
def clean_tp(df):
    t = np.asarray(df["t"], float)
    p = np.asarray(df["p"], float)
    m = np.isfinite(t) & np.isfinite(p) & (t > 0)
    t, p = t[m], p[m]
    idx = np.argsort(t)
    return t[idx], p[idx]

# ---------------------------
# Estimar tc(N) = mximo de dp/dlog t
# ---------------------------
def estimate_tc(df):
    t, p = clean_tp(df)
    logt = np.log(t)
    dp_dlogt = np.gradient(p, logt)
    k = np.argmax(dp_dlogt)
    return t[k]

# ---------------------------
# Score de colapso em janela fsica
# ---------------------------
def collapse_score(dfs, Ns, theta, tcs, mode="pre", ngrid=400):
    curves = []

    for df, N, tc in zip(dfs, Ns, tcs):
        t, p = clean_tp(df)

        if mode == "pre":
            mask = t <= tc
        else:
            mask = t >= tc

        if np.sum(mask) < 10:
            continue

        t, p = t[mask], p[mask]
        x = t * (N ** (-theta))
        idx = np.argsort(x)
        curves.append((x[idx], p[idx]))

    if len(curves) < 2:
        return np.inf

    left = max(c[0][0] for c in curves)
    right = min(c[0][-1] for c in curves)
    if not (left < right):
        return np.inf

    grid = np.linspace(left, right, ngrid)
    P = np.vstack([np.interp(grid, x, p) for x, p in curves])

    return float(np.mean(np.var(P, axis=0)))

# ---------------------------
# Busca de theta por colapso
# ---------------------------
def estimate_theta_by_collapse(dfs, Ns, tcs, mode="pre",
                              theta_min=-5, theta_max=5,
                              npoints=401):

    grid = np.linspace(theta_min, theta_max, npoints)
    scores = np.array([
        collapse_score(dfs, Ns, th, tcs, mode=mode)
        for th in grid
    ])

    theta_star = float(grid[np.argmin(scores)])
    return theta_star, grid, scores

# ---------------------------
# Pipeline principal para C=N
# ---------------------------
def analyze_CN(mode="random"):
    set_fallback_cm_fonts()

    Ns = np.array([300, 500, 600, 1000, 2000, 3000], float)
    CN_dfs = [load_csv_attack(int(N), int(N), mode) for N in Ns]

    # 1) Estima tc(N)
    tcs = np.array([estimate_tc(df) for df in CN_dfs])

    print("\n=== t_c(N) estimado automaticamente ===")
    for N, tc in zip(Ns, tcs):
        print(f"N = {int(N):4d}  ->  t_c  {tc:.3e}")

    # 2) Estima theta_- (pr-crtico)
    theta_minus, grid_m, scores_m = estimate_theta_by_collapse(
        CN_dfs, Ns, tcs, mode="pre", theta_min=-5, theta_max=5
    )

    # 3) Estima theta_+ (ps-crtico)
    theta_plus, grid_p, scores_p = estimate_theta_by_collapse(
        CN_dfs, Ns, tcs, mode="post", theta_min=-5, theta_max=5
    )

    print("\n=== Exponentes de colapso para C = N ===")
    print(f"theta_- (pr)   {theta_minus:.3f}")
    print(f"theta_+ (ps)   {theta_plus:.3f}")

    # ---------------------------
    # Plots dos colapsos
    # ---------------------------
    fig, ax = plt.subplots(1, 2, figsize=(14, 5), constrained_layout=True)

    # Pr-crtico
    for df, N, tc in zip(CN_dfs, Ns, tcs):
        t, p = clean_tp(df)
        mask = t <= tc
        ax[0].plot(t[mask] * (N ** (-theta_minus)), p[mask], lw=2, label=f"N={int(N)}")

    ax[0].set_title(rf"$C=N$ (pr)  $\theta_- \approx {theta_minus:.2f}$")
    ax[0].set_xscale("log")
    ax[0].set_xlabel(r"$tN^{-\theta_-}$")
    ax[0].set_ylabel(r"$p$")
    ax[0].legend(fontsize=9)

    # Ps-crtico
    for df, N, tc in zip(CN_dfs, Ns, tcs):
        t, p = clean_tp(df)
        mask = t >= tc
        ax[1].plot(t[mask] * (N ** (-theta_plus)), p[mask], lw=2, label=f"N={int(N)}")

    ax[1].set_title(rf"$C=N$ (ps)  $\theta_+ \approx {theta_plus:.2f}$")
    ax[1].set_xscale("log")
    ax[1].set_xlabel(r"$tN^{-\theta_+}$")
    ax[1].set_ylabel(r"$p$")
    ax[1].legend(fontsize=9)

    fig.savefig("collapse_CN_pre_post.png", dpi=300, bbox_inches="tight")
    plt.show()


if __name__ == "__main__":
    analyze_CN(mode="random")
```

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import linregress

# --- sua funo ---
def load_csv_attack(N, C, mode="random"):
    fname = f"./data_random_medias_random/{mode}_{N}_{C}.csv"
    if not os.path.exists(fname):
        print(f"[AVISO] Arquivo no encontrado: {fname}")
        return pd.DataFrame()
    df = pd.read_csv(fname, sep=',', header=0)
    return df.drop(columns=["node1", "node2"], errors="ignore")


# --- utilitrio: achar cruzamento por interpolao ---
def find_crossing(p, P, tau):
    p = np.asarray(p, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(p)
    p, P = p[idx], P[idx]

    for i in range(len(p) - 1):
        if P[i] > tau and P[i + 1] <= tau:
            pa, pb = p[i], p[i + 1]
            Pa, Pb = P[i], P[i + 1]
            return pa + (pb - pa) * (Pa - tau) / (Pa - Pb)
    return None


# --- calcula p1, p2 e p ---
def compute_delta_p_from_means(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None, None, None

    col_p = "p"     # frao removida
    col_P = "m1"    # maior componente

    if col_p not in df.columns or col_P not in df.columns:
        raise ValueError(f"Colunas no encontradas. Disponveis: {list(df.columns)}")

    p = df[col_p].values
    P = df[col_P].values

    tau2 = 0.5
    tau1 = 1.0 / np.sqrt(N)

    if debug:
        print(f"[DEBUG] N={N}, C={C}: m1  [{P.min():.3f}, {P.max():.3f}], "
              f"limiares: 0.5 e {tau1:.3f}")

    p2 = find_crossing(p, P, tau2)
    p1 = find_crossing(p, P, tau1)

    if p2 is None:
        print(f"[AVISO] m1 no cruza 0.5 para N={N}, C={C}")
    if p1 is None:
        print(f"[AVISO] m1 no cruza 1/sqrt(N)={tau1:.3f} para N={N}, C={C}")

    if p1 is None or p2 is None:
        return None, None, None

    return p1, p2, (p1 - p2)


# --- ajuste de lei de potncia para extrair alpha ---
def fit_alpha(Ns, deltas):
    Ns = np.asarray(Ns, dtype=float)
    deltas = np.asarray(deltas, dtype=float)

    mask = (deltas > 0) & np.isfinite(deltas)
    Ns, deltas = Ns[mask], deltas[mask]

    if len(Ns) < 2:
        print("[ERRO] Pontos insuficientes para ajustar alpha (precisa de pelo menos 2 Ns vlidos).")
        return {
            "alpha": np.nan,
            "alpha_err": np.nan,
            "r2": 0.0,
            "slope": np.nan,
            "intercept": np.nan
        }

    x = np.log(Ns)
    y = np.log(deltas)

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    alpha = -slope

    return {
        "alpha": alpha,
        "alpha_err": std_err,
        "r2": r_value**2,
        "slope": slope,
        "intercept": intercept
    }


# --- EXECUO ---
if __name__ == "__main__":
    Ns = [600, 1000, 2000, 3000]
    C = 3            # depois rode com C = N para ver o caso abrupto
    mode = "random"

    Ns_ok = []
    deltas = []

    for N in Ns:
        p1, p2, dp = compute_delta_p_from_means(N, C, mode=mode, debug=True)
        if dp is not None:
            Ns_ok.append(N)
            deltas.append(dp)
            print(f"[OK] N={N}: p2={p2:.6f}, p1={p1:.6f}, p={dp:.6e}")

    res = fit_alpha(Ns_ok, deltas)

    print("\n=== Resultado do fit ===")
    print(f"alpha = {res['alpha']:.3f}  {res['alpha_err']:.3f}")
    print(f"r^2   = {res['r2']:.3f}")
```

## Metodologia: Quantificao da Abrupticidade da Transio no SPP

Para quantificar a abrupticidade da transio de percolao no modelo de *Shortest-Path Percolation* (SPP), analisamos como o tamanho do maior componente conectado decresce  medida que a frao de arestas removidas aumenta. Seja $P(p)$ a frao de ns pertencentes ao maior componente conectado em funo da frao de arestas removidas $p$.

A largura da janela de transio  definida como:

$$
\Delta p(N) = p_1(N) - p_2(N),
$$

onde:

- $p_2(N)$  o maior valor de $p$ tal que $P(p) > 0.5$, caracterizando o regime em que a rede ainda possui um componente gigante;
- $p_1(N)$  o maior valor de $p$ tal que $P(p) > 1/\sqrt{N}$, indicando que o maior componente j se tornou sublinear no tamanho da rede.

Esses dois limiares delimitam o intervalo em que ocorre o colapso da conectividade global da rede.

Para cada tamanho de rede $N$, a curva mdia $P(p)$  obtida a partir de mltiplas realizaes do processo SPP. Os valores $p_1(N)$ e $p_2(N)$ so estimados por interpolao linear nos pontos em que $P(p)$ cruza os limiares $0.5$ e $1/\sqrt{N}$, respectivamente.

A fim de caracterizar o carter contnuo ou abrupto da transio, analisamos o escalonamento de $\Delta p(N)$ com o tamanho da rede:

$$
\Delta p(N) \sim N^{-\alpha}.
$$

Tomando logaritmos em ambos os lados, obtemos:

$$
\log \Delta p = -\alpha \log N + \text{const}.
$$

O expoente $\alpha$  ento estimado por regresso linear em escala loglog. Valores $\alpha \approx 0$ indicam uma transio contnua, tpica da percolao ordinria, enquanto $\alpha > 0$ caracteriza um colapso abrupto da conectividade, associado  presena de correlaes topolgicas induzidas pela dinmica de caminhos mnimos no modelo SPP.

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import linregress

# =========================================================
# Leitura dos dados
# =========================================================
def load_csv_attack(N, C, mode="random"):
    fname = f"./data_random_medias_random/{mode}_{N}_{C}.csv"
    if not os.path.exists(fname):
        print(f"[AVISO] Arquivo no encontrado: {fname}")
        return pd.DataFrame()
    df = pd.read_csv(fname, sep=',', header=0)
    return df.drop(columns=["node1", "node2"], errors="ignore")


# =========================================================
# Cruzamento por interpolao linear
# =========================================================
def find_crossing(p, P, tau):
    p = np.asarray(p, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(p)
    p, P = p[idx], P[idx]

    for i in range(len(p) - 1):
        if P[i] > tau and P[i + 1] <= tau:
            pa, pb = p[i], p[i + 1]
            Pa, Pb = P[i], P[i + 1]
            return pa + (pb - pa) * (Pa - tau) / (Pa - Pb)
    return None


# =========================================================
# Pseudo-crtico via mxima derivada + rapidez local
# =========================================================
def find_pc_derivative_and_slope(p, P):
    p = np.asarray(p, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(p)
    p, P = p[idx], P[idx]

    dPdp = np.abs(np.gradient(P, p))
    i_max = np.argmax(dPdp)

    return p[i_max], dPdp[i_max]


# =========================================================
# Computa pseudo-crticos, p e mtricas de rapidez
# =========================================================
def compute_quantities(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None

    p = df["p"].values
    P = df["m1"].values

    # pseudo-crticos
    tau_half = 0.5
    tau_small = 1.0 / np.sqrt(N)

    p_half = find_crossing(p, P, tau_half)
    p_small = find_crossing(p, P, tau_small)
    p_deriv, max_slope = find_pc_derivative_and_slope(p, P)

    # largura global da queda (rapidez "global")
    p_075 = find_crossing(p, P, 0.75)
    p_025 = find_crossing(p, P, 0.25)
    width_7525 = None if (p_075 is None or p_025 is None) else (p_025 - p_075)

    if debug:
        print(f"[DEBUG] N={N}, C={C}")
        print(f"   p(m1=0.5)        = {p_half}")
        print(f"   p(m1=1/sqrtN)   = {p_small}")
        print(f"   p(max|dm1/dp|)  = {p_deriv}")
        print(f"   max|dm1/dp|     = {max_slope}")
        print(f"   width(0.750.25)= {width_7525}")

    if p_half is None or p_small is None or p_deriv is None:
        return None

    delta_p = p_small - p_half

    return {
        "N": N,
        "C": C,
        "p_half": p_half,
        "p_small": p_small,
        "p_deriv": p_deriv,
        "delta_p": delta_p,
        "max_slope": max_slope,
        "width_7525": width_7525
    }


# =========================================================
# Ajuste de p ~ N^{-alpha} (para criticalidade)
# =========================================================
def fit_alpha(Ns, deltas):
    Ns = np.asarray(Ns, dtype=float)
    deltas = np.asarray(deltas, dtype=float)

    mask = deltas > 0
    Ns, deltas = Ns[mask], deltas[mask]

    if len(Ns) < 2:
        return {"alpha": np.nan, "alpha_err": np.nan, "r2": 0.0}

    x = np.log(Ns)
    y = np.log(deltas)

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    alpha = -slope

    return {"alpha": alpha, "alpha_err": std_err, "r2": r_value**2}


# =========================================================
# EXECUO PRINCIPAL
# =========================================================
if __name__ == "__main__":

    Ns = [600, 1000, 2000, 3000]
    mode = "random"

    for C_case in ["C=1", "C=N"]:
        print("\n" + "="*60)
        print(f"CASO {C_case}")
        print("="*60)

        results = []

        for N in Ns:
            C = 1 if C_case == "C=1" else N
            out = compute_quantities(N, C, mode=mode, debug=True)
            if out is not None:
                results.append(out)

        print("\n=== Tabela ===")
        for r in results:
            print(
                f"N={r['N']:4d} | "
                f"p_half={r['p_half']:.5f} | "
                f"p_small={r['p_small']:.5f} | "
                f"p_deriv={r['p_deriv']:.5f} | "
                f"p={r['delta_p']:.5e} | "
                f"max|dm1/dp|={r['max_slope']:.5e} | "
                f"w(0.750.25)={r['width_7525']}"
            )

        # Fit FSS s para p (criticalidade)
        Ns_fit = [r["N"] for r in results]
        deltas = [r["delta_p"] for r in results]
        res = fit_alpha(Ns_fit, deltas)

        print("\n=== Fit p ~ N^{-alpha} (criticalidade) ===")
        print(f"alpha = {res['alpha']:.3f}  {res['alpha_err']:.3f}")
        print(f"r^2   = {res['r2']:.3f}")

        print("\n=== Diagnstico ===")
        if res["alpha"] < 0.1 or np.isnan(res["alpha"]):
            print(" NO crtico (pseudo-crticos no convergem)")
        else:
            print(" Crtico (pseudo-crticos convergem)")

        print(" Rapidez local (maior inclinao): compare max|dm1/dp|")
        print(" Rapidez global (queda 75%25%): compare w(0.750.25)")
```

## O que  criticalidade no sentido de FSS?

FSS (*finite-size scaling*)  a hiptese de que existe um nico ponto crtico $p_c$ tal que, quando o tamanho da rede $N$ cresce:

- As curvas $\langle P(p, N) \rangle$ (ou $\langle m_1(p, N) \rangle$) para diferentes $N$ colapsam quando reescaladas por

  $$
  x = (p - p_c) N^{1/\bar{\nu}}.
  $$

- O pseudo-crtico definido por algum critrio (por exemplo)

  $$
  p_c(N) = \arg\max_p \langle S(p, N) \rangle
  $$

  converge para $p_c$ segundo

  $$
  p_c(N) = p_c + b N^{-1/\bar{\nu}}.
  $$

Consequentemente, qualquer definio razovel de pseudo-crtico (por exemplo, o valor de $p$ onde $m_1(p, N) = 0.5$, onde $|dm_1/dp|$  mximo, etc.) deve convergir para o mesmo valor $p_c$ quando $N \to \infty$.

---

##  O que quer dizer no apresenta criticalidade no sentido de FSS?

Dizer isso  afirmar que essas propriedades falham. No seu caso concreto (para $C = N$):

-  Os pseudo-crticos no colapsam:

  $$
  p_{0.5}(N) \not\to p_{1/\sqrt{N}}(N),
  $$

  pois a diferena

  $$
  \Delta p(N) = p_{1/\sqrt{N}}(N) - p_{0.5}(N)
  $$

  no vai a zero quando $N$ cresce.

-  Logo, no existe um nico $p_c$ que organize a transio.

-  Portanto, o ansatz de FSS

  $$
  \langle P(p, N) \rangle
  =
  N^{-\beta/\bar{\nu}}
  F\!\left[(p - p_c) N^{1/\bar{\nu}}\right]
  $$

  no descreve o comportamento observado para esse regime.

Em linguagem simples, mas precisa:

> No existe um ponto de transio universal que, ao aumentar o tamanho da rede, concentre toda a mudana de conectividade em torno de um nico valor de $p$.

---

##  Traduo em portugus claro

Voc pode substituir a frase tcnica por algo assim:

- **Verso tcnica curta:**  
  o regime $C = N$ no  governado por um ponto crtico nico no sentido de *finite-size scaling*.

- **Verso sem jargo:**  
  ao aumentar o tamanho da rede, no surge um limiar nico que concentre a quebra da conectividade.

- **Verso operacional (ligada aos seus dados):**  
  diferentes critrios para definir o limiar de fragmentao no convergem para o mesmo valor quando $N$ cresce.

---

###  Em uma linha (pra concluso do artigo)

> Para $C = N$, os pseudo-crticos definidos por diferentes critrios permanecem separados com o aumento de $N$, o que  incompatvel com o cenrio de *finite-size scaling* e indica a ausncia de um ponto crtico nico que governe a transio.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.lines import Line2D

mpl.rcParams['text.usetex'] = False

# ---------------------------
# Limpeza bsica
# ---------------------------
def clean_tp(df):
    t = np.asarray(df["t"], float)
    p = np.asarray(df["p"], float)
    m = np.isfinite(t) & np.isfinite(p) & (t > 0)
    t, p = t[m], p[m]
    idx = np.argsort(t)
    return t[idx], p[idx]

# ---------------------------
# Detecta tc(N) via mximo de dp/dlogt
# ---------------------------
def estimate_tc_by_max_slope(df):
    t, p = clean_tp(df)
    if len(t) < 20:
        return np.nan
    logt = np.log(t)
    dp_dlogt = np.gradient(p, logt)
    dp_dlogt = np.convolve(dp_dlogt, np.ones(7)/7, mode="same")
    return t[np.argmax(dp_dlogt)]

# ---------------------------
# Score de colapso fsico
# ---------------------------
def collapse_score_physical(dfs, Ns, theta, masks=None, pmin=0.02, pmax=0.9, ngrid=400):
    curves = []
    for i, (df, N) in enumerate(zip(dfs, Ns)):
        t, p = clean_tp(df)

        mask = masks[i] if masks is not None else (p >= pmin) & (p <= pmax)
        if np.sum(mask) < 10:
            continue

        t, p = t[mask], p[mask]
        x = t * (N ** (-theta))
        idx = np.argsort(x)
        curves.append((x[idx], p[idx]))

    if len(curves) < 2:
        return np.inf

    left = max(c[0][0] for c in curves)
    right = min(c[0][-1] for c in curves)
    if not left < right:
        return np.inf

    grid = np.linspace(left, right, ngrid)
    P = np.vstack([np.interp(grid, x, p) for x, p in curves])

    dPdx = np.gradient(np.mean(P, axis=0), grid)
    weight = np.abs(dPdx) + 1e-7

    return float(np.average(np.var(P, axis=0), weights=weight))

# ---------------------------
# Busca de theta por colapso
# ---------------------------
def estimate_theta_by_collapse(dfs, Ns, masks=None,
                              theta_min=0.0, theta_max=6.0,
                              coarse_points=200, refine_points=200,
                              pmin=0.05, pmax=0.9):

    grid1 = np.linspace(theta_min, theta_max, coarse_points)
    scores1 = np.array([
        collapse_score_physical(dfs, Ns, th, masks=masks, pmin=pmin, pmax=pmax)
        for th in grid1
    ])
    th1 = float(grid1[np.argmin(scores1)])

    step = (theta_max - theta_min) / (coarse_points - 1)
    left = max(theta_min, th1 - 3 * step)
    right = min(theta_max, th1 + 3 * step)

    grid2 = np.linspace(left, right, refine_points)
    scores2 = np.array([
        collapse_score_physical(dfs, Ns, th, masks=masks, pmin=pmin, pmax=pmax)
        for th in grid2
    ])
    th2 = float(grid2[np.argmin(scores2)])

    return th2

# ---------------------------
# Plot final 2x2
# ---------------------------
def plot_final(mode="random"):
    Ns = np.array([300, 500, 600, 1000, 2000, 3000], float)

    C1_dfs = [load_csv_attack(int(N), 1, mode) for N in Ns]
    C2_dfs = [load_csv_attack(int(N), 2, mode) for N in Ns]
    C3_dfs = [load_csv_attack(int(N), 3, mode) for N in Ns]
    CN_dfs = [load_csv_attack(int(N), int(N), mode) for N in Ns]

    theta_C1 = estimate_theta_by_collapse(C1_dfs, Ns)
    theta_C2 = estimate_theta_by_collapse(C2_dfs, Ns)
    theta_C3 = estimate_theta_by_collapse(C3_dfs, Ns)

    tcs_CN = np.array([estimate_tc_by_max_slope(df) for df in CN_dfs])
    post_masks = []
    for df, tc in zip(CN_dfs, tcs_CN):
        t, _ = clean_tp(df)
        post_masks.append(t > tc)

    theta_CN_plus = estimate_theta_by_collapse(CN_dfs, Ns, masks=post_masks)

    cmap = plt.cm.get_cmap('winter')
    styles = [
        {'color': cmap(1.0),  'linestyle': '-',  'linewidth': 2.},
        {'color': cmap(0.8),  'linestyle': '--', 'linewidth': 2.},
        {'color': cmap(0.6),  'linestyle': ':',  'linewidth': 2.},
        {'color': cmap(0.4),  'linestyle': '-.', 'linewidth': 2.},
        {'color': cmap(0.2),  'linestyle': (0, (3, 3)), 'linewidth': 2.},
        {'color': cmap(0.0),  'linestyle': (0, (1, 3)), 'linewidth': 2.}
    ]

    fig, ax = plt.subplots(2, 2, figsize=(14, 10), constrained_layout=True)

    for a in ax.flat:
        a.set_xscale("log")
        a.set_ylabel(r"$P$")
        a.tick_params(direction="in", top=True, right=True)
        a.margins(x=0)

    # -------- (0,0) C=1 --------
    for (df, N, st) in zip(C1_dfs, Ns, styles):
        t, p = clean_tp(df)
        ax[0,0].plot(t * N**(-theta_C1), p, **st)

    # Legenda "caixa nica": C=1, theta, depois N
    handles = [
        Line2D([0], [0], color='none', lw=0, label=r"$C=1$"),
        Line2D([0], [0], color='none', lw=0, label=rf"$\theta \approx {theta_C1:.2f}$"),
    ]
    for N, st in zip(Ns, styles):
        handles.append(Line2D([0], [0],
                              color=st['color'],
                              linestyle=st['linestyle'],
                              linewidth=2.8,
                              label=rf"$N={int(N)}$"))

    ax[0,0].legend(handles=handles,
                   loc='upper left',
                   fontsize=10,
                   frameon=True,
                   fancybox=True,
                   edgecolor='black')

    ax[0,0].set_xlabel(r"$tN^{-\theta}$")

    # -------- (0,1) C=2 --------
    for (df, N, st) in zip(C2_dfs, Ns, styles):
        t, p = clean_tp(df)
        ax[0,1].plot(t * N**(-theta_C2), p, **st)

    ax[0,1].set_xlabel(r"$tN^{-\theta}$")
    ax[0,1].text(0.05, 0.95,
                 rf"$C=2$" + "\n" + rf"$\theta \approx {theta_C2:.2f}$",
                 transform=ax[0,1].transAxes,
                 fontsize=11, va="top",
                 bbox=dict(boxstyle="round", fc="white", ec="black"))

    # -------- (1,0) C=3 --------
    for (df, N, st) in zip(C3_dfs, Ns, styles):
        t, p = clean_tp(df)
        ax[1,0].plot(t * N**(-theta_C3), p, **st)

    ax[1,0].set_xlabel(r"$tN^{-\theta}$")
    ax[1,0].text(0.05, 0.95,
                 rf"$C=3$" + "\n" + rf"$\theta \approx {theta_C3:.2f}$",
                 transform=ax[1,0].transAxes,
                 fontsize=11, va="top",
                 bbox=dict(boxstyle="round", fc="white", ec="black"))

    # -------- (1,1) C=N --------
    for (df, N, st) in zip(CN_dfs, Ns, styles):
        t, p = clean_tp(df)
        ax[1,1].plot(t * N**(-theta_CN_plus), p, **st)

    ax[1,1].set_xlabel(r"$tN^{-\theta^+}$")
    ax[1,1].text(0.05, 0.95,
                 rf"$C=N$" + "\n" + rf"$\theta^+ \approx {theta_CN_plus:.2f}$",
                 transform=ax[1,1].transAxes,
                 fontsize=11, va="top",
                 bbox=dict(boxstyle="round", fc="white", ec="black"))

    fig.savefig("collapse_C1_C2_C3_CN_theta_plus.png", dpi=300, bbox_inches="tight")
    plt.show()


if __name__ == "__main__":
    plot_final(mode="random")
```

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import linregress

# =========================================================
# Leitura dos dados
# =========================================================
def load_csv_attack(N, C, mode="random"):
    fname = f"./data_random_medias_random/{mode}_{N}_{C}.csv"
    if not os.path.exists(fname):
        print(f"[AVISO] Arquivo no encontrado: {fname}")
        return pd.DataFrame()
    df = pd.read_csv(fname, sep=',', header=0)
    return df.drop(columns=["node1", "node2"], errors="ignore")

# =========================================================
# Cruzamento por interpolao linear
# =========================================================
def find_crossing(x, P, tau):
    x = np.asarray(x, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(x)
    x, P = x[idx], P[idx]

    for i in range(len(x) - 1):
        if P[i] > tau and P[i + 1] <= tau:
            xa, xb = x[i], x[i + 1]
            Pa, Pb = P[i], P[i + 1]
            return xa + (xb - xa) * (Pa - tau) / (Pa - Pb)
    return None

# =========================================================
# Pseudo-crtico via mxima derivada (no eixo p)
# =========================================================
def find_pc_derivative(p, P):
    p = np.asarray(p, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(p)
    p, P = p[idx], P[idx]

    dPdp = np.abs(np.gradient(P, p))
    i_max = np.argmax(dPdp)

    return p[i_max]

# =========================================================
# Computa p
# =========================================================
def compute_delta_p(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None

    p = df["p"].values
    P = df["m1"].values

    tau_half = 0.5
    tau_small = 1.0 / np.sqrt(N)

    p_half  = find_crossing(p, P, tau_half)
    p_small = find_crossing(p, P, tau_small)
    p_deriv = find_pc_derivative(p, P)

    if debug:
        print(f"[DEBUG-p] N={N}, C={C}")
        print(f"   p(m1=0.5)        = {p_half}")
        print(f"   p(m1=1/sqrtN)   = {p_small}")
        print(f"   p(max|dm1/dp|)  = {p_deriv}")

    if p_half is None or p_small is None:
        return None

    return p_small - p_half

# =========================================================
# Computa t~
# =========================================================
def compute_delta_ttilde(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None

    t = df["t"].values
    P = df["m1"].values

    tau_half = 0.5
    tau_small = 1.0 / np.sqrt(N)

    t_half  = find_crossing(t, P, tau_half)
    t_small = find_crossing(t, P, tau_small)

    if debug:
        print(f"[DEBUG-t] N={N}, C={C}")
        print(f"   t(m1=0.5)        = {t_half}")
        print(f"   t(m1=1/sqrtN)   = {t_small}")

    if t_half is None or t_small is None:
        return None

    # normalizao como no paper (opcional)
    return (t_small - t_half) / (N**2)

# =========================================================
# Ajuste de lei de potncia
# =========================================================
def fit_alpha(Ns, deltas):
    Ns = np.asarray(Ns, dtype=float)
    deltas = np.asarray(deltas, dtype=float)

    mask = deltas > 0
    Ns, deltas = Ns[mask], deltas[mask]

    if len(Ns) < 2:
        return {"alpha": np.nan, "alpha_err": np.nan, "r2": 0.0}

    x = np.log(Ns)
    y = np.log(deltas)

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    alpha = -slope

    return {"alpha": alpha, "alpha_err": std_err, "r2": r_value**2}

# =========================================================
# EXECUO PRINCIPAL
# =========================================================
if __name__ == "__main__":

    Ns = [600, 1000, 2000, 3000]
    mode = "random"

    for C_case in ["C=1", "C=N"]:
        print("\n" + "="*60)
        print(f"CASO {C_case}")
        print("="*60)

        deltas_p = []
        deltas_t = []
        Ns_ok_p = []
        Ns_ok_t = []

        for N in Ns:
            C = 1 if C_case == "C=1" else N

            dp = compute_delta_p(N, C, mode=mode, debug=True)
            if dp is not None and dp > 0:
                deltas_p.append(dp)
                Ns_ok_p.append(N)

            dt = compute_delta_ttilde(N, C, mode=mode, debug=True)
            if dt is not None and dt > 0:
                deltas_t.append(dt)
                Ns_ok_t.append(N)

        res_p = fit_alpha(Ns_ok_p, deltas_p)
        res_t = fit_alpha(Ns_ok_t, deltas_t)

        print("\n=== Fit p ~ N^{-alpha} ===")
        print(f"alpha  = {res_p['alpha']:.3f}  {res_p['alpha_err']:.3f}")
        print(f"r^2    = {res_p['r2']:.3f}")

        print("\n=== Fit t~ ~ N^{-alpha'} ===")
        print(f"alpha' = {res_t['alpha']:.3f}  {res_t['alpha_err']:.3f}")
        print(f"r^2    = {res_t['r2']:.3f}")
```

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import linregress

# =========================================================
# Leitura dos dados
# =========================================================
def load_csv_attack(N, C, mode="random"):
    fname = f"./data_random_medias_random/{mode}_{N}_{C}.csv"
    if not os.path.exists(fname):
        print(f"[AVISO] Arquivo no encontrado: {fname}")
        return pd.DataFrame()
    df = pd.read_csv(fname, sep=',', header=0)
    return df.drop(columns=["node1", "node2"], errors="ignore")

# =========================================================
# p_critrio: maior x tal que P(x) > tau
# =========================================================
def find_last_above_threshold(x, P, tau):
    x = np.asarray(x, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(x)
    x, P = x[idx], P[idx]

    mask = P > tau
    if not np.any(mask):
        return None

    return x[mask][-1]   # maior x tal que P > tau

# =========================================================
# Computa p = p1 - p2 (como no paper)
# =========================================================
def compute_delta_p(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None

    p = df["p"].values
    P = df["m1"].values

    tau2 = 0.5
    tau1 = 1.0 / np.sqrt(N)

    p2 = find_last_above_threshold(p, P, tau2)
    p1 = find_last_above_threshold(p, P, tau1)

    if debug:
        print(f"[DEBUG-p] N={N}, C={C}")
        print(f"   p2 (P > 0.5)        = {p2}")
        print(f"   p1 (P > 1/sqrtN)   = {p1}")

    if p1 is None or p2 is None:
        return None

    return p1 - p2

# =========================================================
# Computa t~ = (t1 - t2) / N^2
# =========================================================
def compute_delta_ttilde(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None

    t = df["t"].values
    P = df["m1"].values

    tau2 = 0.5
    tau1 = 1.0 / np.sqrt(N)

    t2 = find_last_above_threshold(t, P, tau2)
    t1 = find_last_above_threshold(t, P, tau1)

    if debug:
        print(f"[DEBUG-t] N={N}, C={C}")
        print(f"   t2 (P > 0.5)        = {t2}")
        print(f"   t1 (P > 1/sqrtN)   = {t1}")

    if t1 is None or t2 is None:
        return None

    return (t1 - t2) / (N**2)

# =========================================================
# Ajuste de lei de potncia
# =========================================================
def fit_alpha(Ns, deltas):
    Ns = np.asarray(Ns, dtype=float)
    deltas = np.asarray(deltas, dtype=float)

    mask = deltas > 0
    Ns, deltas = Ns[mask], deltas[mask]

    if len(Ns) < 2:
        return {"alpha": np.nan, "alpha_err": np.nan, "r2": 0.0}

    x = np.log(Ns)
    y = np.log(deltas)

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    alpha = -slope

    return {"alpha": alpha, "alpha_err": std_err, "r2": r_value**2}

# =========================================================
# EXECUO PRINCIPAL
# =========================================================
if __name__ == "__main__":

    Ns = [600, 1000, 2000, 3000]
    mode = "random"

    C_cases = {
        "C=1": lambda N: 1,
        "C=2": lambda N: 2,
        "C=3": lambda N: 3,
        "C=N": lambda N: N,
    }

    print("\n" + "="*70)
    print(" ESTIMATIVA DOS EXPOENTES alpha e alpha' (Kim 2024 style) ")
    print("="*70)

    for label, C_fun in C_cases.items():
        print("\n" + "-"*70)
        print(f"CASO {label}")
        print("-"*70)

        deltas_p, Ns_ok_p = [], []
        deltas_t, Ns_ok_t = [], []

        for N in Ns:
            C = C_fun(N)

            dp = compute_delta_p(N, C, mode=mode, debug=True)
            if dp is not None and dp > 0:
                deltas_p.append(dp)
                Ns_ok_p.append(N)

            dt = compute_delta_ttilde(N, C, mode=mode, debug=True)
            if dt is not None and dt > 0:
                deltas_t.append(dt)
                Ns_ok_t.append(N)

        res_p = fit_alpha(Ns_ok_p, deltas_p)
        res_t = fit_alpha(Ns_ok_t, deltas_t)

        print("\np(N) ~ N^{-alpha}")
        print(f"  alpha  = {res_p['alpha']:.3f}  {res_p['alpha_err']:.3f}")
        print(f"  r^2    = {res_p['r2']:.3f}")

        print("\nt~(N) ~ N^{-alpha'}")
        print(f"  alpha' = {res_t['alpha']:.3f}  {res_t['alpha_err']:.3f}")
        print(f"  r^2    = {res_t['r2']:.3f}")
```

```python
import os
import numpy as np
import pandas as pd
from scipy.stats import linregress

# =========================================================
# Leitura dos dados
# =========================================================
def load_csv_attack(N, C, mode="random"):
    fname = f"./data_random_medias_random/{mode}_{N}_{C}.csv"
    if not os.path.exists(fname):
        print(f"[AVISO] Arquivo no encontrado: {fname}")
        return pd.DataFrame()
    df = pd.read_csv(fname, sep=',', header=0)
    return df.drop(columns=["node1", "node2"], errors="ignore")

# =========================================================
# maior x tal que P(x) > tau
# =========================================================
def find_last_above_threshold(x, P, tau):
    x = np.asarray(x, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(x)
    x, P = x[idx], P[idx]

    mask = P > tau
    if not np.any(mask):
        return None

    return x[mask][-1]

# =========================================================
# pseudo-crtico via mxima derivada |dm1/dp|
# =========================================================
def find_pc_derivative(p, P):
    p = np.asarray(p, dtype=float)
    P = np.asarray(P, dtype=float)

    idx = np.argsort(p)
    p, P = p[idx], P[idx]

    # gradiente numrico
    dPdp = np.abs(np.gradient(P, p))
    i_max = np.argmax(dPdp)

    return p[i_max]

# =========================================================
# Computa p = p1 - p2 (como no paper)
# =========================================================
def compute_delta_p(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None

    p = df["p"].values
    P = df["m1"].values

    tau2 = 0.5
    tau1 = 1.0 / np.sqrt(N)

    p2 = find_last_above_threshold(p, P, tau2)
    p1 = find_last_above_threshold(p, P, tau1)

    if debug:
        print(f"[DEBUG-p] N={N}, C={C}")
        print(f"   p2 (P > 0.5)        = {p2}")
        print(f"   p1 (P > 1/sqrtN)   = {p1}")

    if p1 is None or p2 is None:
        return None

    return p1 - p2

# =========================================================
# Computa t~ = (t1 - t2) / N^2
# =========================================================
def compute_delta_ttilde(N, C, mode="random", debug=True):
    df = load_csv_attack(N, C, mode=mode)
    if df.empty:
        return None

    t = df["t"].values
    P = df["m1"].values

    tau2 = 0.5
    tau1 = 1.0 / np.sqrt(N)

    t2 = find_last_above_threshold(t, P, tau2)
    t1 = find_last_above_threshold(t, P, tau1)

    if debug:
        print(f"[DEBUG-t] N={N}, C={C}")
        print(f"   t2 (P > 0.5)        = {t2}")
        print(f"   t1 (P > 1/sqrtN)   = {t1}")

    if t1 is None or t2 is None:
        return None

    return (t1 - t2) / (N**2)

# =========================================================
# Ajuste de lei de potncia
# =========================================================
def fit_alpha(Ns, deltas):
    Ns = np.asarray(Ns, dtype=float)
    deltas = np.asarray(deltas, dtype=float)

    mask = deltas > 0
    Ns, deltas = Ns[mask], deltas[mask]

    if len(Ns) < 2:
        return {"alpha": np.nan, "alpha_err": np.nan, "r2": 0.0}

    x = np.log(Ns)
    y = np.log(deltas)

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    alpha = -slope

    return {"alpha": alpha, "alpha_err": std_err, "r2": r_value**2}

# =========================================================
# EXECUO PRINCIPAL
# =========================================================
if __name__ == "__main__":

    Ns = [600, 1000, 2000, 3000]
    mode = "random"

    C_cases = {
        "C=1": lambda N: 1,
        "C=2": lambda N: 2,
        "C=3": lambda N: 3,
        "C=N": lambda N: N,
    }

    print("\n" + "="*80)
    print(" ESTIMATIVA DE p_c(N), alpha e alpha' ")
    print("="*80)

    for label, C_fun in C_cases.items():
        print("\n" + "-"*80)
        print(f"CASO {label}")
        print("-"*80)

        deltas_p, Ns_ok_p = [], []
        deltas_t, Ns_ok_t = [], []

        for N in Ns:
            C = C_fun(N)

            df = load_csv_attack(N, C, mode=mode)
            if df.empty:
                continue

            p = df["p"].values
            P = df["m1"].values

            # p_c por dois critrios
            pc_half = find_last_above_threshold(p, P, tau=0.5)
            pc_der  = find_pc_derivative(p, P)

            print(f"[pc] N={N}, C={C}: p_c(half) = {pc_half}, p_c(deriv) = {pc_der}")

            dp = compute_delta_p(N, C, mode=mode, debug=False)
            if dp is not None and dp > 0:
                deltas_p.append(dp)
                Ns_ok_p.append(N)

            dt = compute_delta_ttilde(N, C, mode=mode, debug=False)
            if dt is not None and dt > 0:
                deltas_t.append(dt)
                Ns_ok_t.append(N)

        res_p = fit_alpha(Ns_ok_p, deltas_p)
        res_t = fit_alpha(Ns_ok_t, deltas_t)

        print("\np(N) ~ N^{-alpha}")
        print(f"  alpha  = {res_p['alpha']:.3f}  {res_p['alpha_err']:.3f}")
        print(f"  r^2    = {res_p['r2']:.3f}")

        print("\nt~(N) ~ N^{-alpha'}")
        print(f"  alpha' = {res_t['alpha']:.3f}  {res_t['alpha_err']:.3f}")
        print(f"  r^2    = {res_t['r2']:.3f}")
```
